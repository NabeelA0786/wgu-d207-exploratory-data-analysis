```python
import pandas as pd
from scipy.stats import ttest_ind

# Load the dataset
file_path = "churn_clean.csv"
data = pd.read_csv(file_path)

# Split the data into churned and non-churned groups
churn_yes = data[data['Churn'] == 'Yes']['MonthlyCharge']
churn_no = data[data['Churn'] == 'No']['MonthlyCharge']

# Run the t-test
t_stat, p_value = ttest_ind(churn_yes, churn_no, equal_var=False)

# Output the results
print(f"T-Statistic: {round(t_stat, 3)}")
print(f"P-Value: {p_value:.3e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "churn_clean.csv"
data = pd.read_csv(file_path)

# Continuous Variables - Histograms
plt.figure(figsize=(10, 5))
plt.hist(data['MonthlyCharge'], bins=30, color='blue', edgecolor='black')
plt.title('Distribution of Monthly Charge')
plt.xlabel('Monthly Charge')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 5))
plt.hist(data['Tenure'], bins=30, color='green', edgecolor='black')
plt.title('Distribution of Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Frequency')
plt.show()

# Categorical Variables - Bar Charts
data['Churn'].value_counts().plot(kind='bar', color=['orange', 'red'], edgecolor='black', figsize=(8, 5))
plt.title('Distribution of Churn')
plt.xlabel('Churn (Yes/No)')
plt.ylabel('Count')
plt.show()

data['InternetService'].value_counts().plot(kind='bar', color=['purple', 'cyan', 'magenta'], edgecolor='black', figsize=(10, 5))
plt.title('Distribution of Internet Service')
plt.xlabel('Internet Service Type')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "churn_clean.csv"
data = pd.read_csv(file_path)

# Continuous vs. Continuous - Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Tenure'], data['MonthlyCharge'], alpha=0.6, color='blue')
plt.title('Scatter Plot of Monthly Charge vs. Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Monthly Charge')
plt.grid(True)
plt.show()

# Box Plot: MonthlyCharge by Churn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='MonthlyCharge', data=data, hue='Churn', palette='Set2', dodge=False)
plt.title('Monthly Charge Distribution by Churn Status')
plt.xlabel('Churn (Yes/No)')
plt.ylabel('Monthly Charge')
plt.legend(title='Churn')
plt.show()

# Categorical vs. Categorical - Grouped Bar Chart
churn_internet = pd.crosstab(data['Churn'], data['InternetService'])
churn_internet.plot(kind='bar', stacked=True, figsize=(10, 6), color=['purple', 'cyan', 'magenta'], edgecolor='black')
plt.title('Churn by Internet Service')
plt.xlabel('Churn (Yes/No)')
plt.ylabel('Count')
plt.legend(title='Internet Service')
plt.show()
```
